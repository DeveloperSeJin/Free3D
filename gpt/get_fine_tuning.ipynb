```python
import openai

openai.api_key = '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'

def getPrompt(text) :
    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "user", "content" : text},
        ]
    )
    return response.choices[0].message.content
```

```python
question = open('./question2gpt.txt', 'r')

question_text = "".join(q for q in question)
print(question_text)
```

```python
question_list = question_text.split('\n\n')
question_list
```

```python
answer = open('./answer.txt', 'w')

num = 1
for q in question_list :
    a = getPrompt(q)
    answer.write(str(num) + '\n')
    answer.write(a)
    answer.write('\n')
    answer.write('-' * 150)
    answer.write('\n')
    num += 1
    
answer.close()
question.close()
```
